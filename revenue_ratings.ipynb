```python
total_por_lingua = tmdb['original_language'].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc['en']
total_do_resto = total_geral - total_de_ingles
print(total_por_lingua)
print(total_geral)
print(total_de_ingles)
print(total_do_resto)
```

```python
contagem_por_lingua = tmdb['original_language'].value_counts()
contagem_por_lingua
```

```python
dados = {
"lingua" : ["ingles", "outros"],
"total" : [total_de_ingles, total_do_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(data=dados, x='lingua', y='total', palette='mako')
plt.title('DistribuiÃ§Ã£o de filmes por lÃ­ngua')
plt.xlabel('LÃ­ngua')
plt.ylabel('Total')
```

```python
dados.plot(kind='pie', labels=dados['lingua'], y='total')
```
